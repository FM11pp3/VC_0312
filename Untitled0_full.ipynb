# Parte B - Funções usadas para criar o CNN

```python
train_frames = {
    "A": {"num_classes": 1, "base": "A", "use_cgan": False},
    "B": {"num_classes": len(classes_map)-1, "base": "B", "use_cgan": False},
    "C": {"num_classes": len(classes_map), "base": "C", "use_cgan": False},
    "D": {"num_classes": len(classes_map)-1, "base": "B", "use_cgan": True},
    "E": {"num_classes": len(classes_map), "base": "C", "use_cgan": True},
}

def resolve_variant(key: str, use_cgan_balance: bool | None):
    key = key.upper()
    if key not in train_frames:
        raise ValueError(f"Modelo invalido: {key}. Use A, B, C, D ou E.")
    cfg = train_frames[key]
    base_key = cfg["base"]
    num_classes = cfg["num_classes"]
    use_cgan = cfg["use_cgan"] if use_cgan_balance is None else use_cgan_balance
    return base_key, num_classes, use_cgan

def build_train_df(base_df: pd.DataFrame, base_key: str):
    base_key = base_key.upper()
    if base_key == "A":
        df = base_df.assign(label=base_df["class_name"].apply(lambda c: 0 if c == "No-Anomaly" else 1))
    elif base_key == "B":
        df = base_df[base_df["class_name"] != "No-Anomaly"].copy() # Use .copy() to avoid SettingWithCopyWarning
        df["label"] = df["class_name"].map(classes_map_B)
        df = df.dropna(subset=["label"]) # Explicitly remove rows where label mapping resulted in NaN
    elif base_key == "C":
        df = base_df.assign(label=lambda d: d["class_name"].map(classes_map))
    else:
        raise ValueError(f"Base invalida: {base_key}.")
    # Ensure label column is integer type after all processing
    df['label'] = df['label'].astype(int)
    return df.reset_index(drop=True)


def get_base_df(use_cgan_balance: bool, target_per_class=None):
    if not use_cgan_balance:
        return train_df
    if "balance_with_cgan" not in globals():
        raise NameError("balance_with_cgan nao definido. Corre a celula da Parte C primeiro.")
    return balance_with_cgan(target_per_class=target_per_class)


def train_model(
    model_name: str,
    base_df: pd.DataFrame,
    num_classes: int,
    epochs: int = 3,
    lr: float = 1e-3,
    use_weighted_sampler: bool = True,
    batch_size: int = 128,
):
    train_split, val_split = train_test_split(
        base_df,
        test_size=0.2,
        stratify=base_df["label"],
        random_state=SEED,
    )
    sampler = make_weighted_sampler(train_split) if use_weighted_sampler else None
    train_loader = make_loader(
        train_split,
        train_transform,
        batch_size=batch_size,
        shuffle=not use_weighted_sampler,
        sampler=sampler,
    )
    val_loader = make_loader(val_split, test_transform, batch_size=batch_size)

    model = NetworkCNN(num_classes).to(DEVICE)
    optimizer = torch.optim.Adam(model.parameters(), lr=lr)
    is_binary = num_classes == 1
    criterion = nn.BCEWithLogitsLoss() if is_binary else nn.CrossEntropyLoss()

    for epoch in range(epochs):
        model.train()
        running_loss = 0.0
        for images, labels in train_loader:
            images, labels = images.to(DEVICE), labels.to(DEVICE)
            optimizer.zero_grad()
            if is_binary:
                logits = model(images).view(-1)
                loss = criterion(logits, labels.float())
            else:
                loss = criterion(model(images), labels)
            loss.backward()
            optimizer.step()
            running_loss += loss.item() * labels.size(0)
        val_metrics = evaluate_model_metrics(model, val_loader, num_classes, idx_to_classes, plot_results=False, model_name=f"{model_name}_val")
        val_acc = val_metrics["accuracy"]
        print(
            f"Epoch {epoch + 1}/{epochs} | loss={running_loss / len(train_loader.dataset):.4f} | val_acc={val_acc:.3f}"
        )

    out_path = MODELS_DIR / f"{model_name}.pth"
    torch.save(model.state_dict(), out_path)
    print(f"Modelo guardado em {out_path}")
    return model


def train_model_variant(key: str, use_cgan_balance: bool | None = None, target_per_class=None, **kwargs):
    base_key, num_classes, use_cgan = resolve_variant(key, use_cgan_balance)
    base_df = get_base_df(use_cgan, target_per_class=target_per_class)
    df = build_train_df(base_df, base_key)
    name = f"model_{key}_scratch"
    return train_model(name, df, num_classes=num_classes, **kwargs)


def run_kfold_cv(
    model_name: str,
    base_df: pd.DataFrame,
    num_classes: int,
    n_splits: int = 5,
    epochs: int = 3,
    batch_size: int = 128,
    lr: float = 1e-3,
    use_weighted_sampler: bool = True,
):
    skf = StratifiedKFold(n_splits=n_splits, shuffle=True, random_state=SEED)
    fold_rows = []

    for fold_idx, (train_idx, val_idx) in enumerate(skf.split(base_df, base_df["label"]), start=1):
        print(f"\nFold {fold_idx}/{n_splits}")
        train_split = base_df.iloc[train_idx].reset_index(drop=True)
        val_split = base_df.iloc[val_idx].reset_index(drop=True)
        sampler = make_weighted_sampler(train_split) if use_weighted_sampler else None

        train_loader = make_loader(
            train_split,
            train_transform,
            batch_size=batch_size,
            shuffle=not use_weighted_sampler,
            sampler=sampler,
        )
        val_loader = make_loader(val_split, test_transform, batch_size=batch_size)

        model = NetworkCNN(num_classes).to(DEVICE)
        optimizer = torch.optim.Adam(model.parameters(), lr=lr)
        is_binary = num_classes == 1
        criterion = nn.BCEWithLogitsLoss() if is_binary else nn.CrossEntropyLoss()

        for epoch in range(epochs):
            model.train()
            running_loss = 0.0
            for images, labels in train_loader:
                images, labels = images.to(DEVICE), labels.to(DEVICE)
                optimizer.zero_grad()
                if is_binary:
                    logits = model(images).view(-1)
                    loss = criterion(logits, labels.float())
                else:
                    loss = criterion(model(images), labels)
                loss.backward()
                optimizer.step()
                running_loss += loss.item() * labels.size(0)
            val_metrics = evaluate_model_metrics(model, val_loader, num_classes, idx_to_classes, plot_results=False, model_name=f"{model_name}_fold{fold_idx}_val")
            val_acc = val_metrics["accuracy"]
            print(
                f"  Epoch {epoch + 1}/{epochs} | loss={running_loss / len(train_loader.dataset):.4f} | val_acc={val_acc:.3f}"
            )

        metrics = evaluate_model_metrics(model, val_loader, num_classes, idx_to_classes, plot_results=False, model_name=f"{model_name}_fold{fold_idx}_final")
        fold_rows.append({"fold": fold_idx, **metrics})
        print(
            f"Fold {fold_idx} metrics: acc={metrics['accuracy']:.3f} | f1_macro={metrics['f1_macro']:.3f} | f1_weighted={metrics['f1_weighted']:.3f}"
        )

    fold_df = pd.DataFrame(fold_rows)
    metrics_path = METRICS_DIR / f"kfold_results_{model_name}.csv"
    METRICS_DIR.mkdir(exist_ok=True)
    fold_df.to_csv(metrics_path, index=False)
    print(f"Metricas K-fold guardadas em {metrics_path}")
    display(fold_df)
    return fold_df


def run_kfold_variant(key: str, use_cgan_balance: bool | None = None, target_per_class=None, **kwargs):
    base_key, num_classes, use_cgan = resolve_variant(key, use_cgan_balance)
    base_df = get_base_df(use_cgan, target_per_class=target_per_class)
    df = build_train_df(base_df, base_key)
    # Map idx_to_classes_B for models B and D
    current_idx_to_classes = idx_to_classes
    if base_key == "B":
        current_idx_to_classes = classes_map_B # Use the anomaly classes map
        # Convert classes_map_B to idx_to_classes_B for consistent use
        current_idx_to_classes = {v: k for k, v in current_idx_to_classes.items()}

    return run_kfold_cv(f"{key}", df, num_classes=num_classes, **kwargs, class_labels=current_idx_to_classes)


def run_kfold_all(keys=("A", "B", "C", "D", "E"), target_per_class=None, **kwargs):
    summaries = {}
    for key in keys:
        print(f"\n===== Modelo {key} =====")
        _, num_classes, use_cgan_default = resolve_variant(key, use_cgan_balance=None)
        # Map idx_to_classes_B for models B and D
        current_idx_to_classes = idx_to_classes
        if key in ("B", "D"):
            current_idx_to_classes = {v: k for k, v in classes_map_B.items()}

        summaries[key] = run_kfold_variant(key, use_cgan_balance=use_cgan_default, target_per_class=target_per_class, **kwargs, class_labels=current_idx_to_classes)
    return summaries


# Exemplos (comentados):
# balanced_train_df = get_base_df(use_cgan_balance=True)  # gera e mistura reais+GAN
# run_kfold_variant("A", n_splits=5, epochs=3)            # Binario: real
# run_kfold_variant("D", n_splits=5, epochs=3)            # 11 classes anomalas: real + CGAN
# run_kfold_variant("B", n_splits=5, epochs=3)            # 11 classes anomalas: real
# run_kfold_variant("C", n_splits=5, epochs=3)            # 12 classes: real
# run_kfold_variant("E", n_splits=5, epochs=3)            # 12 classes: real + CGAN
# train_model_variant("E", epochs=5)                      # Treino rapido modelo E com dados mix


# ===============================
# Pack de treino A-E com SGD+StepLR, early stopping e 20 epochs por defeito
# ===============================
from torch.optim.lr_scheduler import StepLR


def make_optimizer_and_sched(model, lr=1e-2, momentum=0.9, weight_decay=1e-4, step_size=7, gamma=0.1):
    opt = torch.optim.SGD(model.parameters(), lr=lr, momentum=momentum, weight_decay=weight_decay, nesterov=True)
    sched = StepLR(opt, step_size=step_size, gamma=gamma)
    return opt, sched


def train_model_with_es(
    model_name: str,
    base_df: pd.DataFrame,
    num_classes: int,
    epochs: int = 20,
    lr: float = 1e-2,
    momentum: float = 0.9,
    weight_decay: float = 1e-4,
    step_size: int = 7,
    gamma: float = 0.1,
    patience: int = 4,
    batch_size: int = 128,
    use_weighted_sampler: bool = True,
):
    """Treina com SGD+StepLR e early stopping (melhor f1_macro). Guarda o melhor estado."""
    train_split, val_split = train_test_split(
        base_df, test_size=0.2, stratify=base_df["label"], random_state=SEED
    )
    sampler = make_weighted_sampler(train_split) if use_weighted_sampler else None
    train_loader = make_loader(train_split, train_transform, batch_size=batch_size, shuffle=not use_weighted_sampler, sampler=sampler)
    val_loader = make_loader(val_split, test_transform, batch_size=batch_size)

    model = NetworkCNN(num_classes).to(DEVICE)
    is_binary = num_classes == 1
    criterion = nn.BCEWithLogitsLoss() if is_binary else nn.CrossEntropyLoss()
    optimizer, scheduler = make_optimizer_and_sched(model, lr=lr, momentum=momentum, weight_decay=weight_decay, step_size=step_size, gamma=gamma)

    best_f1 = -1.0
    best_state = None
    patience_ctr = 0

    # Determine current_idx_to_classes_train for this model_name
    current_idx_to_classes_train = idx_to_classes # Default to all 12 classes
    key_for_map = model_name.split('_')[1] # Extracts 'A', 'B', 'C', 'D', 'E'
    if key_for_map in ("B", "D"):
        current_idx_to_classes_train = {v: k for k, v in classes_map_B.items()}

    for epoch in range(epochs):
        model.train()
        running_loss = 0.0
        for images, labels in train_loader:
            images, labels = images.to(DEVICE), labels.to(DEVICE)
            optimizer.zero_grad()
            if is_binary:
                logits = model(images).view(-1)
                loss = criterion(logits, labels.float())
            else:
                loss = criterion(model(images), labels)
            loss.backward()
            optimizer.step()
            running_loss += loss.item() * labels.size(0)
        scheduler.step()

        val_metrics = evaluate_model_metrics(model, val_loader, num_classes, current_idx_to_classes_train, plot_results=False, model_name=f"{model_name}_val_epoch{epoch+1}")
        val_f1 = val_metrics["f1_macro"]
        print(f"{model_name} | epoch {epoch+1}/{epochs} | loss={running_loss/len(train_loader.dataset):.4f} | val_f1={val_f1:.3f}")

        if val_f1 > best_f1 + 1e-4:
            best_f1 = val_f1
            best_state = model.state_dict()
            patience_ctr = 0
        else:
            patience_ctr += 1
            if patience_ctr >= patience:
                print(f"Early stopping em epoch {epoch+1} (sem melhoria {patience} épocas)")
                break

    if best_state is None:
        best_state = model.state_dict()
    out_path = MODELS_DIR / f"{model_name}.pth"
    torch.save(best_state, out_path)
    print(f"Modelo guardado em {out_path} com f1_macro={best_f1:.3f}")
    # Pass the correct class_labels for the final metrics call
    final_metrics = evaluate_model_metrics(model, val_loader, num_classes, current_idx_to_classes_train, plot_results=False, model_name=f"{model_name}_final")
    print(f"Final metrics for {model_name}: Acc={final_metrics['accuracy']:.3f} | F1={final_metrics['f1_macro']:.3f}")
    return best_state, best_f1


def run_kfold_with_es(
    model_name: str,
    base_df: pd.DataFrame,
    num_classes: int,
    n_splits: int = 5,
    epochs: int = 20,
    **kwargs,
):
    """K-fold com SGD+StepLR e early stopping. Devolve métricas por fold."""
    skf = StratifiedKFold(n_splits=n_splits, shuffle=True, random_state=SEED)
    fold_rows = []
    for fold_idx, (train_idx, val_idx) in enumerate(skf.split(base_df, base_df["label"]), start=1):
        print(f"\nFold {fold_idx}/{n_splits}")
        train_split = base_df.iloc[train_idx].reset_index(drop=True)
        val_split = base_df.iloc[val_idx].reset_index(drop=True)
        sampler = make_weighted_sampler(train_split) if kwargs.get("use_weighted_sampler", True) else None
        train_loader = make_loader(train_split, train_transform, batch_size=kwargs.get("batch_size", 128), shuffle=not kwargs.get("use_weighted_sampler", True), sampler=sampler)
        val_loader = make_loader(val_split, test_transform, batch_size=kwargs.get("batch_size", 128))

        model = NetworkCNN(num_classes).to(DEVICE)
        is_binary = num_classes == 1
        criterion = nn.BCEWithLogitsLoss() if is_binary else nn.CrossEntropyLoss()
        optimizer, scheduler = make_optimizer_and_sched(model, lr=kwargs.get("lr", 1e-2), momentum=kwargs.get("momentum", 0.9), weight_decay=kwargs.get("weight_decay", 1e-4), step_size=kwargs.get("step_size", 7), gamma=kwargs.get("gamma", 0.1))

        best_f1 = -1.0
        patience = kwargs.get("patience", 4)
        patience_ctr = 0
        best_state = None

        # Determine current_idx_to_classes for this model_name
        current_idx_to_classes_fold = idx_to_classes # Default to all 12 classes
        key_for_map = model_name # model_name is already 'A', 'B', 'C', 'D', 'E'
        if key_for_map in ("B", "D"):
            current_idx_to_classes_fold = {v: k for k, v in classes_map_B.items()}

        for epoch in range(epochs):
            model.train()
            running_loss = 0.0
            for images, labels in train_loader:
                images, labels = images.to(DEVICE), labels.to(DEVICE)
                optimizer.zero_grad()
                if is_binary:
                    logits = model(images).view(-1)
                    loss = criterion(logits, labels.float())
                else:
                    loss = criterion(model(images), labels)
                loss.backward()
                optimizer.step()
                running_loss += loss.item() * labels.size(0)
            scheduler.step()

            val_metrics = evaluate_model_metrics(model, val_loader, num_classes, current_idx_to_classes_fold, plot_results=False, model_name=f"{model_name}_fold{fold_idx}_val_epoch{epoch+1}")
            val_f1 = val_metrics["f1_macro"]
            print(f"  Epoch {epoch+1}/{epochs} | loss={running_loss/len(train_loader.dataset):.4f} | val_f1={val_f1:.3f}")
            if val_f1 > best_f1 + 1e-4:
                best_f1 = val_f1
                best_state = model.state_dict()
                patience_ctr = 0
            else:
                patience_ctr += 1
                if patience_ctr >= patience:
                    print(f"  Early stopping neste fold (sem melhoria {patience} épocas)")
                    break

        if best_state is None:
            best_state = model.state_dict()
        metrics = evaluate_model_metrics(model, val_loader, num_classes, current_idx_to_classes_fold, plot_results=False, model_name=f"{model_name}_fold{fold_idx}_final")
        fold_rows.append({"fold": fold_idx, **metrics, "best_f1_macro": best_f1})

    fold_df = pd.DataFrame(fold_rows)
    out_path = METRICS_DIR / f"kfold_results_es_{model_name}.csv"
    METRICS_DIR.mkdir(exist_ok=True)
    fold_df.to_csv(out_path, index=False)
    print(f"Métricas K-fold guardadas em {out_path}")
    return fold_df


def train_all_variants_with_es(
    keys=("A", "B", "C", "D", "E"),
    use_kfold: bool = True,
    epochs: int = 20,
    **kwargs,
):
    """Treina variantes A-E com SGD+StepLR e early stopping. Guarda pesos em models/model_<key>_es.pth"""
    results = {}
    ensure_dir(RESULTS_BASE)

    for key in keys:
        print(f"\n🚀 Variante {key}")

        base_key, num_classes, use_cgan_default = resolve_variant(key, use_cgan_balance=None)
        use_cgan = kwargs.get("use_cgan_balance", use_cgan_default)
        base_df = get_base_df(use_cgan, target_per_class=kwargs.get("target_per_class"))
        df = build_train_df(base_df, base_key)

        if use_kfold:
            print(f"\n=== K-fold {key} ===")
            results[f"kfold_{key}"] = run_kfold_with_es(f"{key}", df, num_classes=num_classes, epochs=epochs, **kwargs)

        print(f"\n=== Treino final {key} ===")
        best_state, best_f1 = train_model_with_es(
            model_name=f"model_{key}_es",
            base_df=df,
            num_classes=num_classes,
            epochs=epochs,
            **kwargs,
        )
        results[f"model_{key}"] = {"best_f1_macro": best_f1}
    return results

# Exemplo de uso rápido (corre esta célula após definir balance_with_cgan se quiser usar D/E):
# train_all_variants_with_es(keys=("A","B","C","D","E"), use_kfold=True, epochs=20, lr=1e-2, momentum=0.9, patience=4)
```

```python
# Dataset, modelo e helpers para avaliacao
class SolarDataset(Dataset):
    def __init__(self, df: pd.DataFrame, transform):
        self.df = df.reset_index(drop=True)
        self.transform = transform

    def __len__(self):
        return len(self.df)

    def __getitem__(self, idx):
        row = self.df.iloc[idx]
        image = Image.open(row["path"]).convert("L")
        if self.transform:
            image = self.transform(image)
        return image, int(row["label"])


def make_loader(df, transform, batch_size=256, shuffle=False, sampler=None):
    return DataLoader(
        SolarDataset(df, transform),
        batch_size=batch_size,
        shuffle=shuffle if sampler is None else False,
        sampler=sampler,
        num_workers=2,
        pin_memory=torch.cuda.is_available(),
    )


class NetworkCNN(nn.Module):
    def __init__(self, num_classes: int):
        super().__init__()
        self.conv1 = nn.Conv2d(1, 6, 5)
        self.pool = nn.MaxPool2d(2, 2)
        self.conv2 = nn.Conv2d(6, 16, 5)

        dummy_input = torch.randn(1, 1, 64, 64)
        with torch.no_grad():
            x = self.pool(F.relu(self.conv1(dummy_input)))
            x = self.pool(F.relu(self.conv2(x)))
            flattened_size = torch.flatten(x, 1).shape[1]

        self.fc1 = nn.Linear(flattened_size, 120)
        self.fc2 = nn.Linear(120, 84)
        self.fc3 = nn.Linear(84, num_classes)

    def forward(self, x):
        x = self.pool(F.relu(self.conv1(x)))
        x = self.pool(F.relu(self.conv2(x)))
        x = torch.flatten(x, 1)
        x = F.relu(self.fc1(x))
        x = F.relu(self.fc2(x))
        x = self.fc3(x)
        return x



def predict_classes(logits: torch.Tensor) -> torch.Tensor:
    """Converte sa?das do modelo em r?tulos.
    - 1 logit: sigmoid + threshold 0.5
    - >1 logits: argmax
    """
    if logits.ndim == 1:
        logits = logits.unsqueeze(1)
    if logits.shape[1] == 1:
        probs = torch.sigmoid(logits)
        return (probs >= 0.5).long().view(-1)
    return logits.argmax(1)
```

```python
from sklearn.metrics import accuracy_score, f1_score, precision_score, recall_score, roc_auc_score, confusion_matrix, RocCurveDisplay
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np

def evaluate_model_metrics(model: nn.Module, loader: DataLoader, num_classes: int, class_labels: dict, plot_results: bool = False, model_name: str = ""):
    model.eval()
    all_preds, all_labels, all_logits = [], [], []
    with torch.no_grad():
        for images, labels in loader:
            images, labels = images.to(DEVICE), labels.to(DEVICE)
            logits = model(images)
            preds = predict_classes(logits) # predict_classes is already defined

            all_logits.extend(logits.cpu().tolist())
            all_preds.extend(preds.cpu().tolist())
            all_labels.extend(labels.cpu().tolist())

    if not all_labels:
        return {
            "accuracy": 0.0, "f1_macro": 0.0, "f1_micro": 0.0, "f1_weighted": 0.0,
            "precision_macro": 0.0, "recall_macro": 0.0,
            "precision_weighted": 0.0, "recall_weighted": 0.0,
            "model_parameters": sum(p.numel() for p in model.parameters() if p.requires_grad)
        }

    metrics = {
        "accuracy": accuracy_score(all_labels, all_preds),
        "f1_macro": f1_score(all_labels, all_preds, average="macro", zero_division=0),
        "f1_micro": f1_score(all_labels, all_preds, average="micro", zero_division=0),
        "f1_weighted": f1_score(all_labels, all_preds, average="weighted", zero_division=0),
        "precision_macro": precision_score(all_labels, all_preds, average="macro", zero_division=0),
        "recall_macro": recall_score(all_labels, all_preds, average="macro", zero_division=0),
        "precision_weighted": precision_score(all_labels, all_preds, average="weighted", zero_division=0),
        "recall_weighted": recall_score(all_labels, all_preds, average="weighted", zero_division=0),
        "model_parameters": sum(p.numel() for p in model.parameters() if p.requires_grad)
    }

    # ROC AUC
    all_logits_tensor = torch.tensor(all_logits)
    if num_classes == 1: # Binary classification
        all_probs = torch.sigmoid(all_logits_tensor).numpy()
        try:
            metrics["roc_auc"] = roc_auc_score(all_labels, all_probs)
        except ValueError:
            metrics["roc_auc"] = 0.0 # Handle cases with only one class present
        if plot_results:
            plt.figure(figsize=(6, 5))
            RocCurveDisplay.from_predictions(all_labels, all_probs, name=f"ROC {model_name}")
            plt.title(f"ROC Curve - {model_name}")
            plt.show()
    elif num_classes > 1: # Multi-class classification (One-vs-Rest ROC AUC)
        all_probs = torch.softmax(all_logits_tensor, dim=1).numpy()
        try:
            # Use OVR for multi-class AUC; averaging can be macro or weighted
            metrics["roc_auc_macro"] = roc_auc_score(all_labels, all_probs, multi_class='ovr', average='macro')
            metrics["roc_auc_weighted"] = roc_auc_score(all_labels, all_probs, multi_class='ovr', average='weighted')
        except ValueError:
            metrics["roc_auc_macro"] = 0.0
            metrics["roc_auc_weighted"] = 0.0
        # Plotting multi-class ROCs (one curve per class) can be very busy; skipping individual plots by default
        # If plot_results: # This block is commented out to avoid excessive plots for multi-class
        #    for i, class_name in class_labels.items():
        #        if i == 7: continue # Skip No-Anomaly for multi-class plots if desired
        #        fpr, tpr, _ = roc_curve((np.array(all_labels) == i).astype(int), all_probs[:, i])
        #        plt.plot(fpr, tpr, label=f'ROC curve (area = {roc_auc_score((np.array(all_labels) == i).astype(int), all_probs[:, i]):0.2f}) for class {class_name}')
        #    plt.plot([0, 1], [0, 1], 'k--') # Random guess line
        #    plt.xlabel('False Positive Rate')
        #    plt.ylabel('True Positive Rate')
        #    plt.title(f'Multi-class ROC Curves - {model_name}')
        #    plt.legend(loc="lower right")
        #    plt.show()

    # Confusion Matrix
    if plot_results:
        cm = confusion_matrix(all_labels, all_preds)

        # Use a sensible figure size, adapting to number of classes
        fig_size = max(6, num_classes * 0.8) # Minimum 6, scale with num_classes
        plt.figure(figsize=(fig_size, fig_size))

        # Get class names for labels, map integers to strings
        display_labels = [class_labels[i] for i in sorted(class_labels.keys())]
        if num_classes == 1: # For binary, labels are 0 and 1
            display_labels = ['No Anomaly', 'Anomaly'] # Or whatever 0 and 1 mean
        elif len(display_labels) != num_classes:
            display_labels = [str(i) for i in range(num_classes)] # Fallback if map incomplete

        sns.heatmap(cm, annot=True, fmt='d', cmap='Blues',
                    xticklabels=display_labels,
                    yticklabels=display_labels)
        plt.xlabel('Predicted')
        plt.ylabel('True')
        plt.title(f'Confusion Matrix - {model_name}')
        plt.show()

    return metrics


def evaluate_model(model: nn.Module, loader: DataLoader) -> float:
    model.eval()
    correct = total = 0
    with torch.no_grad():
        for images, labels in loader:
            images, labels = images.to(DEVICE), labels.to(DEVICE)
            preds = predict_classes(model(images))
            correct += (preds == labels).sum().item()
            total += labels.size(0)
    return correct / total if total else 0.0


def make_weighted_sampler(df: pd.DataFrame) -> WeightedRandomSampler:
    class_counts = df["label"].value_counts()
    class_weights = 1.0 / class_counts
    sample_weights = df["label"].map(class_weights).astype(float)
    return WeightedRandomSampler(
        weights=torch.as_tensor(sample_weights.values, dtype=torch.double),
        num_samples=len(sample_weights),
        replacement=True,
    )
```

```python
# GAN condicional simples (G e D totalmente conectados) para gerar imagens 64x64 em tons de cinza
LATENT_DIM = 128
NUM_CLASSES = len(classes_map)
GAN_OUT_DIR = REPO_ROOT / "cgan_generated_outputs"
GAN_OUT_DIR.mkdir(exist_ok=True)

class ConditionalGenerator(nn.Module):
    def __init__(self, latent_dim: int, num_classes: int):
        super().__init__()
        self.label_emb = nn.Embedding(num_classes, num_classes)
        self.model = nn.Sequential(
            nn.Linear(latent_dim + num_classes, 256),
            nn.BatchNorm1d(256),
            nn.LeakyReLU(0.2, inplace=True),
            nn.Linear(256, 512),
            nn.BatchNorm1d(512),
            nn.LeakyReLU(0.2, inplace=True),
            nn.Linear(512, 1024),
            nn.BatchNorm1d(1024),
            nn.LeakyReLU(0.2, inplace=True),
            nn.Linear(1024, 1 * 64 * 64),
            nn.Tanh(),
        )

    def forward(self, noise, labels):
        x = torch.cat((self.label_emb(labels), noise), dim=1)
        out = self.model(x)
        return out.view(noise.size(0), 1, 64, 64)

class ConditionalDiscriminator(nn.Module):
    def __init__(self, num_classes: int):
        super().__init__()
        self.label_emb = nn.Embedding(num_classes, num_classes)
        self.model = nn.Sequential(
            nn.Linear(1 * 64 * 64 + num_classes, 512),
            nn.LeakyReLU(0.2, inplace=True),
            nn.Linear(512, 256),
            nn.LeakyReLU(0.2, inplace=True),
            nn.Linear(256, 1),
            nn.Sigmoid(),
        )

    def forward(self, img, labels):
        x = torch.cat((img.view(img.size(0), -1), self.label_emb(labels)), dim=1)
        validity = self.model(x)
        return validity

def make_gan_loader(df, batch_size=64):
    return DataLoader(
        SolarDataset(df, train_transform),
        batch_size=batch_size,
        shuffle=True,
        drop_last=True,
        num_workers=2,
        pin_memory=torch.cuda.is_available(),
    )

def train_cgan(train_df: pd.DataFrame, num_epochs: int = 5, batch_size: int = 64, lr: float = 2e-4, sample_every: int = 1):
    loader = make_gan_loader(train_df, batch_size=batch_size)
    generator = ConditionalGenerator(LATENT_DIM, NUM_CLASSES).to(DEVICE)
    discriminator = ConditionalDiscriminator(NUM_CLASSES).to(DEVICE)
    adversarial_loss = nn.BCELoss()

    opt_G = torch.optim.Adam(generator.parameters(), lr=lr, betas=(0.5, 0.999))
    opt_D = torch.optim.Adam(discriminator.parameters(), lr=lr, betas=(0.5, 0.999))

    fixed_noise = torch.randn(NUM_CLASSES, LATENT_DIM, device=DEVICE)
    fixed_labels = torch.arange(NUM_CLASSES, device=DEVICE)

    for epoch in range(num_epochs):
        for imgs, labels in loader:
            imgs, labels = imgs.to(DEVICE), labels.to(DEVICE)
            valid = torch.ones(imgs.size(0), 1, device=DEVICE)
            fake = torch.zeros(imgs.size(0), 1, device=DEVICE)

            # Train generator
            opt_G.zero_grad()
            z = torch.randn(imgs.size(0), LATENT_DIM, device=DEVICE)
            gen_labels = labels
            gen_imgs = generator(z, gen_labels)
            g_loss = adversarial_loss(discriminator(gen_imgs, gen_labels), valid)
            g_loss.backward()
            opt_G.step()

            # Train discriminator
            opt_D.zero_grad()
            real_loss = adversarial_loss(discriminator(imgs, labels), valid)
            fake_loss = adversarial_loss(discriminator(gen_imgs.detach(), gen_labels), fake)
            d_loss = (real_loss + fake_loss) / 2
            d_loss.backward()
            opt_D.step()

        print(f"Epoch {epoch+1}/{num_epochs} | D loss: {d_loss.item():.4f} | G loss: {g_loss.item():.4f}")

        if (epoch + 1) % sample_every == 0:
            with torch.no_grad():
                samples = generator(fixed_noise, fixed_labels).cpu()
                grid = vutils.make_grid(samples, nrow=4, normalize=True)
                out_path = GAN_OUT_DIR / f"cgan_samples_epoch_{epoch+1:03}.png"
                vutils.save_image(grid, out_path)
                print(f"Samples guardados em {out_path}")

    torch.save(generator.state_dict(), MODELS_DIR / "cgan_generator.pth")
    print("Gerador guardado em models/cgan_generator.pth")
    return generator

# Exemplo de uso (comentado): treinar GAN condicional usando o split de treino completo
# generator = train_cgan(train_df, num_epochs=5, batch_size=128, sample_every=1)
```

```python
# CGAN: classe geradora + funcoes de balanceamento (antes no anexo, agora na Parte C)
# Reconstrução da arquitetura do gerador para corresponder ao código de treino DCGAN (célula 14qC6x7cY3sg)
class ConditionalGenerator(nn.Module):
    def __init__(self, latent_dim: int, num_classes: int):
        super().__init__()
        # Corresponde a nn.Embedding(num_classes, embed_size=100) no código de treino
        self.label_embedding = nn.Embedding(num_classes, 100)

        # Re-criação da sequência principal (self.main) exatamente como no código de treino,
        # garantindo bias=False para todas as ConvTranspose2d e os canais corretos.
        # img_channels é 1 para grayscale.
        self.main = nn.Sequential(
            # Input para esta camada é (batch, latent_dim + 100, 1, 1)
            nn.ConvTranspose2d(latent_dim + 100, 256, 4, 1, 0, bias=False),
            nn.BatchNorm2d(256),
            nn.ReLU(True),
            nn.ConvTranspose2d(256, 128, 4, 2, 1, bias=False),
            nn.BatchNorm2d(128),
            nn.ReLU(True),
            nn.ConvTranspose2d(128, 64, 4, 2, 1, bias=False),
            nn.BatchNorm2d(64),
            nn.ReLU(True),
            nn.ConvTranspose2d(64, 32, 4, 2, 1, bias=False),
            nn.BatchNorm2d(32),
            nn.ReLU(True),
            nn.ConvTranspose2d(32, 1, 4, 2, 1, bias=False), # img_channels = 1
            nn.Tanh()
        )

    def forward(self, noise, labels):
        # noise é (batch, LATENT_DIM) - precisa ser reshape para (batch, LATENT_DIM, 1, 1)
        # labels é (batch,) - label_embedding vai converter para (batch, 100)
        # label_embed precisa ser reshape para (batch, 100, 1, 1)
        label_embed = self.label_embedding(labels).view(-1, 100, 1, 1)
        noise_reshaped = noise.view(-1, LATENT_DIM, 1, 1) # Reshape noise para 4D

        # Concatena noise e embedding da label ao longo da dimensão dos canais
        input_vec = torch.cat([noise_reshaped, label_embed], 1)

        # Passa pelo gerador principal
        return self.main(input_vec)


def load_trained_generator(weights_path=CGAN_WEIGHTS):
    if not weights_path.exists():
        raise FileNotFoundError(f"Nao encontrei pesos do CGAN em {weights_path}. Treina o CGAN ou ajusta o caminho.")
    generator = ConditionalGenerator(LATENT_DIM, NUM_CLASSES).to(DEVICE)
    state = torch.load(weights_path, map_location=DEVICE)
    generator.load_state_dict(state)
    generator.eval()
    return generator


TIER_THRESHOLDS = (500, 1000, 1500)


def tier_target(count: int, thresholds=TIER_THRESHOLDS) -> int:
    for t in thresholds:
        if count < t:
            return t
    return count


def generate_class_samples(generator, class_id: int, num_images: int, batch_size: int = 64):
    saved_paths = []
    remaining = num_images
    GAN_AUG_DIR.mkdir(parents=True, exist_ok=True)
    while remaining > 0:
        cur = min(batch_size, remaining)
        noise = torch.randn(cur, LATENT_DIM, device=DEVICE)
        labels = torch.full((cur,), class_id, device=DEVICE, dtype=torch.long)
        with torch.no_grad():
            imgs = generator(noise, labels).cpu()
        for i in range(cur):
            out_name = f"cgan_{class_id:02d}_{len(saved_paths)+1:05d}.png"
            out_path = GAN_AUG_DIR / out_name
            vutils.save_image(imgs[i], out_path, normalize=True)
            saved_paths.append(out_path)
        remaining -= cur
    return saved_paths


def balance_with_cgan(target_per_class: int | None = None, batch_size: int = 64):
    """Gera imagens sinteticas para cada classe ate target_per_class (default = max count atual)."""
    generator = load_trained_generator()
    class_counts = train_df["label"].value_counts().to_dict()
    if target_per_class is None:
        target_per_class = max(class_counts.values())
    new_rows = []
    for cls_name, cls_id in classes_map.items():
        cur = class_counts.get(cls_id, 0)
        if cur >= target_per_class:
            continue
        need = target_per_class - cur
        print(f"Classe {cls_name} (id={cls_id}): gerar {need} para equilibrar {cur}->{target_per_class}")
        paths = generate_class_samples(generator, cls_id, need, batch_size=batch_size)
        for p in paths:
            new_rows.append({"path": p, "class_name": cls_name, "label": cls_id})
    if not new_rows:
        print("Dataset ja esta balanceado; nada gerado.")
        return train_df
    synth_df = pd.DataFrame(new_rows)
    balanced_df = pd.concat([train_df, synth_df], ignore_index=True)
    print(f"Geradas {len(synth_df)} imagens sinteticas. Total final: {len(balanced_df)}")
    return balanced_df


def balance_with_cgan_tiers(thresholds=TIER_THRESHOLDS, batch_size: int = 64, save_summary: bool = True):
    """Balanceia por patamares (ex.: <500->500, <1000->1000, <1500->1500) e devolve df balanceado + resumo."""
    generator = load_trained_generator()
    class_counts = train_df["label"].value_counts().to_dict()
    new_rows = []
    summary_rows = []
    for cls_name, cls_id in classes_map.items():
        real = class_counts.get(cls_id, 0)
        target = tier_target(real, thresholds=thresholds)
        need = target - real
        synth_paths = []
        if need > 0:
            print(f"Classe {cls_name} (id={cls_id}): gerar {need} para equilibrar {real}->{target}")
            synth_paths = generate_class_samples(generator, cls_id, need, batch_size=batch_size)
            for p in synth_paths:
                new_rows.append({"path": p, "class_name": cls_name, "label": cls_id})
        total = real + len(synth_paths)
        summary_rows.append({
            "class_name": cls_name,
            "label": cls_id,
            "real_count": real,
            "target": target,
            "synth_count": len(synth_paths),
            "total": total,
            "synth_ratio": (len(synth_paths) / total) if total else 0.0,
        })
    synth_df = pd.DataFrame(new_rows)
    balanced_df = pd.concat([train_df, synth_df], ignore_index=True) if len(synth_df) else train_df
    summary_df = pd.DataFrame(summary_rows)
    if save_summary:
        METRICS_DIR.mkdir(parents=True, exist_ok=True)
        out_path = METRICS_DIR / "cgan_tiered_summary.csv"
        summary_df.to_csv(out_path, index=False)
        print(f"Resumo por patamares guardado em: {out_path}")
    print(f"Total sinteticas geradas: {summary_df['synth_count'].sum()}")
    return balanced_df, summary_df


# Exemplos (comentados):
# balanced_train_df = balance_with_cgan()                     # balanceia ate ao maximo atual por classe
balanced_train_df, summary_df = balance_with_cgan_tiers()   # balanceia por patamares 500/1000/1500 e grava resumo
# sampler = make_weighted_sampler(balanced_train_df)
# loader = make_loader(balanced_train_df, train_transform, batch_size=128, sampler=sampler)
```

```python
# --- Prepare cGAN Dataset and DataLoader ---
# 1. Identify the minority classes
minority_class_indices = [idx for cls, idx in classes_map.items() if idx != 7]
minority_class_names = [idx_to_classes[idx] for idx in minority_class_indices]

# 2. Filter full_train_df to create minority_classes_df
minority_classes_df = full_train_df[full_train_df['label'].isin(minority_class_indices)].copy()

# 3. Create an instance of SolarDataset for cGAN training
cgan_train_dataset = SolarDataset(minority_classes_df, transform=train_transform)

# 4. Create a PyTorch DataLoader for cGAN training
cgan_train_loader = torch.utils.data.DataLoader(
    cgan_train_dataset,
    batch_size=32,
    shuffle=True
)
print("`cgan_train_loader` (and dependencies) created.")

# --- Re-define Generator and Discriminator classes ---

# Define cGAN Conditional Generator Network
class ConditionalGenerator(nn.Module):
    def __init__(self, z_dim, img_channels, num_classes, embed_size=100):
        super(ConditionalGenerator, self).__init__()
        self.num_classes = num_classes
        self.img_channels = img_channels
        self.z_dim = z_dim
        self.embed_size = embed_size

        self.label_embedding = nn.Embedding(num_classes, embed_size)
        self.input_dim = z_dim + embed_size

        self.main = nn.Sequential(
            nn.ConvTranspose2d(self.input_dim, 256, 4, 1, 0, bias=False),
            nn.BatchNorm2d(256),
            nn.ReLU(True),
            nn.ConvTranspose2d(256, 128, 4, 2, 1, bias=False),
            nn.BatchNorm2d(128),
            nn.ReLU(True),
            nn.ConvTranspose2d(128, 64, 4, 2, 1, bias=False),
            nn.BatchNorm2d(64),
            nn.ReLU(True),
            nn.ConvTranspose2d(64, 32, 4, 2, 1, bias=False),
            nn.BatchNorm2d(32),
            nn.ReLU(True),
            nn.ConvTranspose2d(32, img_channels, 4, 2, 1, bias=False),
            nn.Tanh()
        )

    def forward(self, noise, labels):
        label_embed = self.label_embedding(labels).view(-1, self.embed_size, 1, 1)
        input_vec = torch.cat([noise, label_embed], 1)
        return self.main(input_vec)

# Define cGAN Conditional Discriminator Network
class ConditionalDiscriminator(nn.Module):
    def __init__(self, img_channels, num_classes, embed_size=100):
        super(ConditionalDiscriminator, self).__init__()
        self.num_classes = num_classes
        self.img_channels = img_channels
        self.embed_size = embed_size

        self.label_embedding = nn.Embedding(num_classes, embed_size)

        self.conv_layers = nn.Sequential(
            nn.Conv2d(img_channels, 64, 4, 2, 1, bias=False),
            nn.LeakyReLU(0.2, inplace=True),
            nn.Conv2d(64, 128, 4, 2, 1, bias=False),
            nn.BatchNorm2d(128),
            nn.LeakyReLU(0.2, inplace=True),
            nn.Conv2d(128, 256, 4, 2, 1, bias=False),
            nn.BatchNorm2d(256),
            nn.LeakyReLU(0.2, inplace=True),
            nn.Conv2d(256, 512, 4, 2, 1, bias=False),
            nn.BatchNorm2d(512),
            nn.LeakyReLU(0.2, inplace=True)
        )

        self.fc = nn.Sequential(
            nn.Linear(512 * 4 * 4 + embed_size, 1024),
            nn.LeakyReLU(0.2, inplace=True),
            nn.Linear(1024, 1)
        )

    def forward(self, image, labels):
        img_features = self.conv_layers(image)
        img_features = torch.flatten(img_features, 1)
        label_embed = self.label_embedding(labels)
        combined_features = torch.cat([img_features, label_embed], 1)
        output = self.fc(combined_features)
        return output

# --- Re-define weights_init helper function ---
def weights_init(m):
    classname = m.__class__.__name__
    if classname.find('Conv') != -1:
        nn.init.normal_(m.weight.data, 0.0, 0.02)
    elif classname.find('BatchNorm') != -1:
        nn.init.normal_(m.weight.data, 1.0, 0.02)
        nn.init.constant_(m.bias.data, 0)

# --- Re-define train_cgan function ---
def train_cgan(generator, discriminator, cgan_train_loader, z_dim, num_epochs, device, lr_g, lr_d, beta1, img_channels, num_classes, cgan_model_path="cgan_generator_minority_classes.pth"):
    generator.apply(weights_init)
    discriminator.apply(weights_init)

    criterion = nn.BCEWithLogitsLoss()

    optimizer_D = optim.Adam(discriminator.parameters(), lr=lr_d, betas=(beta1, 0.999))
    optimizer_G = optim.Adam(generator.parameters(), lr=lr_g, betas=(beta1, 0.999))

    generator.to(device)
    discriminator.to(device)

    fixed_noise = torch.randn(64, z_dim, 1, 1, device=device)
    fixed_labels = torch.randint(0, num_classes, (64,), device=device)

    print("Starting cGAN Training Loop...")
    for epoch in range(num_epochs):
        for i, (real_images, labels) in enumerate(cgan_train_loader):
            ############################
            # (1) Update D network: maximize log(D(x | y)) + log(1 - D(G(z | y)))
            ###########################
            discriminator.zero_grad()
            real_images = real_images.to(device)
            labels = labels.to(device)
            b_size = real_images.size(0)

            real_label = torch.full((b_size,), 1.0, dtype=torch.float, device=device)
            fake_label = torch.full((b_size,), 0.0, dtype=torch.float, device=device)

            output_real = discriminator(real_images, labels).view(-1)
            errD_real = criterion(output_real, real_label)
            errD_real.backward()
            D_x = output_real.mean().item()

            noise = torch.randn(b_size, z_dim, 1, 1, device=device)
            fake_labels = labels # Use real labels as condition for generated fakes

            fake_images = generator(noise, fake_labels)
            output_fake = discriminator(fake_images.detach(), fake_labels).view(-1)
            errD_fake = criterion(output_fake, fake_label)
            errD_fake.backward()
            D_G_z1 = output_fake.mean().item()

            errD = errD_real + errD_fake
            optimizer_D.step()

            ############################
            # (2) Update G network: maximize log(D(G(z | y)))
            ###########################
            generator.zero_grad()
            output_fake_for_G = discriminator(fake_images, fake_labels).view(-1)
            errG = criterion(output_fake_for_G, real_label)
            errG.backward()
            D_G_z2 = output_fake_for_G.mean().item()
            optimizer_G.step()

            if i % 50 == 0:
                print(f'Epoch [{epoch+1}/{num_epochs}] Batch [{i}/{len(cgan_train_loader)}]\tLoss_D: {errD.item():.4f}\tLoss_G: {errG.item():.4f}\tD(x): {D_x:.4f}\tD(G(z)): {D_G_z1:.4f} / {D_G_z2:.4f}')

        generator.eval()
        with torch.no_grad():
            fake_images_epoch = generator(fixed_noise, fixed_labels).detach().cpu()
            vutils.save_image(fake_images_epoch, f"cgan_generated_samples_epoch_{epoch+1:03d}.png", normalize=True)
        generator.train()

    print("cGAN Training Complete!")
    torch.save(generator.state_dict(), cgan_model_path)
    print(f"Trained cGAN Generator saved to: {cgan_model_path}")

print("\n=== Training cGAN and Saving Generator ===")

# 1. Initialize Generator and Discriminator
cgan_generator = ConditionalGenerator(z_dim, img_channels, num_classes).to(DEVICE)
cgan_discriminator = ConditionalDiscriminator(img_channels, num_classes).to(DEVICE)

# 2. Train the cGAN
train_cgan(
    generator=cgan_generator,
    discriminator=cgan_discriminator,
    cgan_train_loader=cgan_train_loader,
    z_dim=z_dim,
    num_epochs=num_epochs,
    device=DEVICE,
    lr_g=lr_g,
    lr_d=lr_d,
    beta1=beta1,
    img_channels=img_channels,
    num_classes=num_classes
)

print("\u2714\ufe0f cGAN training completed and generator saved.")
```

```python
import os
import json
import torch
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.metrics import confusion_matrix, roc_auc_score, RocCurveDisplay

RESULTS_BASE = "/content/VC_0312/results"

def ensure_dir(path: str):
    os.makedirs(path, exist_ok=True)

def save_history_csv(history_dict, save_path):
    df_hist = pd.DataFrame(history_dict)
    df_hist.to_csv(save_path, index=False)

def save_metrics_csv(metrics_dict, save_path):
    df = pd.DataFrame([metrics_dict])
    df.to_csv(save_path, index=False)

def save_summary_json(summary_dict, save_path):
    with open(save_path, "w") as f:
        json.dump(summary_dict, f, indent=4)

def save_loss_acc_plot(history_dict, save_path, model_label):
    epochs = history_dict["epoch"]
    plt.figure(figsize=(12,5))
    plt.subplot(1,2,1)
    plt.plot(epochs, history_dict["train_loss"], label="Train Loss")
    plt.plot(epochs, history_dict["val_loss"], label="Val Loss")
    plt.title(f"Loss - {model_label}")
    plt.legend(); plt.grid(True)

    plt.subplot(1,2,2)
    plt.plot(epochs, history_dict["train_acc"], label="Train Acc")
    plt.plot(epochs, history_dict["val_acc"], label="Val Acc")
    plt.title(f"Accuracy - {model_label}")
    plt.legend(); plt.grid(True)

    plt.tight_layout()
    plt.savefig(save_path)
    plt.close()

def save_confusion_matrix(model, loader, class_labels: dict, save_path):
    model.eval()
    all_preds, all_labels = [], []
    with torch.no_grad():
        for images, labels in loader:
            images, labels = images.to(DEVICE), labels.to(DEVICE)
            preds = predict_classes(model(images))
            all_preds.extend(preds.cpu().numpy())
            all_labels.extend(labels.cpu().numpy())

    cm = confusion_matrix(all_labels, all_preds)
    plt.figure(figsize=(10,8))
    sns.heatmap(cm, annot=True, fmt='d', cmap="Blues",
                xticklabels=list(class_labels.values()),
                yticklabels=list(class_labels.values()))
    plt.title("Confusion Matrix")
    plt.savefig(save_path)
    plt.close()

def save_roc_curve(model, loader, num_classes, save_path):
    model.eval()
    all_labels, all_probs = [], []
    with torch.no_grad():
        for images, labels in loader:
            images = images.to(DEVICE)
            logits = model(images)
            probs = torch.softmax(logits, dim=1).cpu().numpy()
            all_probs.extend(probs)
            all_labels.extend(labels.numpy())

    try:
        RocCurveDisplay.from_predictions(all_labels, np.array(all_probs), multi_class='ovr')
        plt.title("ROC Curve")
        plt.savefig(save_path)
        plt.close()
    except:
        pass
```

```python
from sklearn.model_selection import StratifiedKFold
from torch.utils.data import DataLoader
import torch
import torch.nn as nn

SEED = 42

# ============================================================
# ✔ MODELO FINAL COM EARLY STOPPING E SALVAMENTO AUTOMÁTICO
# ============================================================
def train_model_with_es(model_name, base_df, num_classes, epochs, **kwargs):

    print(f"\n=== Treino Final para {model_name} ===")

    # Split final (80% train, 20% val)
    train_df, val_df = train_test_split(base_df, test_size=0.2,
                                        stratify=base_df["label"], random_state=SEED)

    train_loader = make_loader(train_df, train_transform,
                               batch_size=kwargs.get("batch_size", 128),
                               shuffle=True)

    val_loader = make_loader(val_df, test_transform,
                             batch_size=kwargs.get("batch_size", 128))

    model = NetworkCNN(num_classes).to(DEVICE)
    criterion = nn.CrossEntropyLoss()
    optimizer, scheduler = make_optimizer_and_sched(model)

    best_f1 = -1
    patience = kwargs.get("patience", 4)
    patience_ctr = 0
    best_state = None

    history = {"epoch": [], "train_loss": [], "train_acc": [], "val_loss": [], "val_acc": []}

    for epoch in range(epochs):
        model.train()
        correct = 0
        running_loss = 0

        for images, labels in train_loader:
            images, labels = images.to(DEVICE), labels.to(DEVICE)
            optimizer.zero_grad()
            logits = model(images)
            loss = criterion(logits, labels)
            loss.backward()
            optimizer.step()

            running_loss += loss.item() * labels.size(0)
            pred = logits.argmax(1)
            correct += (pred == labels).sum().item()

        scheduler.step()
        train_loss = running_loss / len(train_loader.dataset)
        train_acc = correct / len(train_loader.dataset)

        # Validação
        val_metrics = evaluate_model_metrics(model, val_loader, num_classes, idx_to_classes)
        val_loss = val_metrics.get("val_loss", 0)  # caso não exista
        val_acc = val_metrics["accuracy"]
        val_f1 = val_metrics["f1_macro"]

        print(f"Epoch {epoch+1}/{epochs} "
              f"| Train Loss={train_loss:.4f} Acc={train_acc:.3f} "
              f"| Val Acc={val_acc:.3f} F1={val_f1:.3f}")

        history["epoch"].append(epoch+1)
        history["train_loss"].append(train_loss)
        history["train_acc"].append(train_acc)
        history["val_loss"].append(val_loss)
        history["val_acc"].append(val_acc)

        # Early Stopping
        if val_f1 > best_f1 + 1e-4:
            best_f1 = val_f1
            best_state = model.state_dict()
            patience_ctr = 0
        else:
            patience_ctr += 1
            if patience_ctr >= patience:
                print("⚠️ Early stopping ativado!")
                break

    # Se nunca melhorou nada
    if best_state is None:
        best_state = model.state_dict()

    # Salvar modelo final
    key = model_name.split("_")[1]
    save_path = f"{RESULTS_BASE}/{key}/final_model"
    ensure_dir(save_path)

    torch.save(best_state, f"{save_path}/model_{key}_final.pth")
    save_history_csv(history, f"{save_path}/history_{key}_final.csv")

    final_metrics = evaluate_model_metrics(model, val_loader, num_classes, idx_to_classes)
    save_metrics_csv(final_metrics, f"{save_path}/metrics_{key}_final.csv")
    save_loss_acc_plot(history, f"{save_path}/loss_acc_{key}_final.png", f"Model {key} - Final")
    save_confusion_matrix(model, val_loader, idx_to_classes, f"{save_path}/confusion_matrix_{key}_final.png")
    save_roc_curve(model, val_loader, num_classes, f"{save_path}/roc_curve_{key}_final.png")

    return best_state, best_f1, history


# ============================================================
# ✔ K-FOLD TRAINING COM EARLY STOPPING E REGISTO COMPLETO
# ============================================================
def run_kfold_with_es(model_name, base_df, num_classes, epochs, **kwargs):
    skf = StratifiedKFold(n_splits=5, shuffle=True, random_state=SEED)

    fold_histories = {}
    fold_rows = []

    for fold_idx, (train_idx, val_idx) in enumerate(skf.split(base_df, base_df["label"]), start=1):
        print(f"\n🔄 Fold {fold_idx}/5 — {model_name}")

        train_df = base_df.iloc[train_idx].reset_index(drop=True)
        val_df = base_df.iloc[val_idx].reset_index(drop=True)

        train_loader = make_loader(train_df, train_transform, batch_size=128, shuffle=True)
        val_loader = make_loader(val_df, test_transform, batch_size=128)

        model = NetworkCNN(num_classes).to(DEVICE)
        criterion = nn.CrossEntropyLoss()
        optimizer, scheduler = make_optimizer_and_sched(model)

        best_f1 = -1
        patience = kwargs.get("patience", 4)
        patience_ctr = 0
        best_state = None

        history = {"epoch": [], "train_loss": [], "train_acc": [], "val_loss": [], "val_acc": []}

        for epoch in range(epochs):
            model.train()
            running_loss = 0
            correct = 0

            for images, labels in train_loader:
                images, labels = images.to(DEVICE), labels.to(DEVICE)
                optimizer.zero_grad()
                logits = model(images)
                loss = criterion(logits, labels)
                loss.backward()
                optimizer.step()
                running_loss += loss.item() * labels.size(0)
                correct += (logits.argmax(1) == labels).sum().item()

            scheduler.step()

            train_loss = running_loss / len(train_loader.dataset)
            train_acc = correct / len(train_loader.dataset)

            val_metrics = evaluate_model_metrics(model, val_loader, num_classes, idx_to_classes)
            val_acc = val_metrics["accuracy"]
            val_f1 = val_metrics["f1_macro"]

            print(f"Fold {fold_idx} | Epoch {epoch+1}/{epochs}: "
                  f"train_loss={train_loss:.4f}, train_acc={train_acc:.3f}, val_f1={val_f1:.3f}")

            history["epoch"].append(epoch+1)
            history["train_loss"].append(train_loss)
            history["train_acc"].append(train_acc)
            history["val_loss"].append(0)
            history["val_acc"].append(val_acc)

            if val_f1 > best_f1 + 1e-4:
                best_f1 = val_f1
                best_state = model.state_dict()
                patience_ctr = 0
            else:
                patience_ctr += 1
                if patience_ctr >= patience:
                    print("🚨 Early stopping no fold!")
                    break

        if best_state is None:
            best_state = model.state_dict()

        fold_histories[fold_idx] = history
        fold_rows.append({"fold": fold_idx, "best_f1_macro": best_f1, "best_state": best_state})

    fold_metrics = pd.DataFrame(fold_rows)
    return fold_metrics, fold_histories


# ============================================================
# ✔ EXECUTOR PRINCIPAL — TREINA VARIANTES, SALVA TUDO
# ============================================================
def train_all_variants_with_es(keys=("A","B","C","D","E"), epochs=20, use_kfold=True, **kwargs):
    results = {}
    ensure_dir(RESULTS_BASE)

    for key in keys:
        print(f"\n🚀 Variante {key}")

        base_key, num_classes, use_cgan_default = resolve_variant(key, use_cgan_balance=None)
        use_cgan = kwargs.get("use_cgan_balance", use_cgan_default)
        base_df = get_base_df(use_cgan, target_per_class=kwargs.get("target_per_class"))
        df = build_train_df(base_df, base_key)

        if use_kfold:
            # K-Fold
            kfold_metrics, kfold_histories = run_kfold_with_es(f"{key}", df, num_classes, epochs, **kwargs)
            best_row = kfold_metrics.sort_values("best_f1_macro", ascending=False).iloc[0]
            best_fold = int(best_row["fold"])
            best_history = kfold_histories[best_fold]

            save_dir = f"{RESULTS_BASE}/{key}/fold_{best_fold}_best"
            ensure_dir(save_dir)

            torch.save(best_row["best_state"], f"{save_dir}/model_{key}_fold{best_fold}_best.pth")
            save_history_csv(best_history, f"{save_dir}/history_{key}_fold{best_fold}_best.csv")
            save_loss_acc_plot(best_history, f"{save_dir}/loss_acc_{key}_fold{best_fold}_best.png",
                               f"Model {key} Fold {best_fold} Best")

            results[f"kfold_{key}"] = {"best_fold": best_fold,
                                       "best_f1": best_row["best_f1_macro"]}

        # Treino final
        best_state, best_f1, history_final = train_model_with_es(
            model_name=f"model_{key}_es",
            base_df=df,
            num_classes=num_classes,
            epochs=epochs,
            **kwargs,
        )

        results[f"model_{key}"] = best_f1

    return results



def plot_best_fold(results_dict, key="A"):
    info = results_dict.get(f"kfold_{key}")
    if not info:
        raise ValueError(f"Nenhum resultado de K-fold encontrado para {key}.")
    hist = info["best_history"]
    best_fold = info["best_fold"]
    best_row = info["best_row"]
    epochs = hist["epoch"]
    plt.figure(figsize=(12,5))
    plt.subplot(1,2,1)
    plt.plot(epochs, hist["train_loss"], label="Training Loss", color="b")
    plt.plot(epochs, hist["val_loss"], label="Validation Loss", color="r")
    plt.title(f"Model {key} Fold {best_fold}: Loss")
    plt.xlabel("Epochs")
    plt.ylabel("Loss")
    plt.legend(); plt.grid(True)

    plt.subplot(1,2,2)
    plt.plot(epochs, hist["train_acc"], label="Training Accuracy", color="b")
    plt.plot(epochs, hist["val_acc"], label="Validation Accuracy", color="r")
    plt.title(f"Model {key} Fold {best_fold}: Accuracy")
    plt.xlabel("Epochs")
    plt.ylabel("Accuracy")
    plt.legend(); plt.grid(True)
    plt.tight_layout()
    plt.show()
    print(f"Melhor fold {best_fold} - f1_macro={best_row['best_f1_macro']:.4f}, acc={best_row['accuracy']:.4f}")

```
